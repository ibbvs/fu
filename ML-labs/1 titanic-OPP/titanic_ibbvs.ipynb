````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/ibbvs/jasa/main/ML-labs/data%20sources/titanic.csv")

df.head()
````

````python
df.info()
````

````python
[print(f"{col:>15}: {df[col].nunique():>5} из {df.shape[0]} строк совпадает. Процент {round(df[col].nunique() / df.shape[0], 2) * 100}%") for col in df.columns]
````

````python
df.describe()
````

````python
df = df.drop(columns=["PassengerId", "Name", "Lname", "Cabin", "Embarked", "Ticket"])

df.head()
````

````python
df["Age"].median(), df["Age"].mean()
````

````python
df["Год Рождения"] = df.fillna(df["Age"].median())["Age"].apply(lambda x: 1912 - x).astype(int)

df.head()
````

````python
df = df.rename(columns={"Survived": "Выжил", "Pclass": "Класс", "Sex": "Пол", "SibSp": "Количество родственников", "Parch": "Количество родителей, детей", "Fare": "Стоимость тарифа"})

df = df.drop(columns="Age")

df.head()
````

````python
df.to_csv("answer.csv")
````
