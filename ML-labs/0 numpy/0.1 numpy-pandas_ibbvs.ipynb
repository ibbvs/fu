##### 2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def func(N, el1, d):
    mask = np.arange(el1, el1 + N*d, d)

    return np.diag(mask)

func(7, 5, 0.5)
```

##### 5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.


```python
m1 = np.zeros((5, 5)).astype("int64")

m1[::2, 1::2] = 5
m1[1::2, ::2] = 5

# np.triu() сохранит только элементы, находящиеся выше или на диагонали (включая диагональные элементы)
m2 = np.triu(np.full((5, 5), 5))


m1, m2
```

```python
det1 = np.linalg.det(m1)
det2 = np.linalg.det(m2)

print(f"детерминант первой матрицы:\n{m1}\n равен {det1}")
```

```python
try:
  inv1 = np.linalg.inv(m1)
  inv1
except:
  print(f"Невозможно посчитать обратную матрицу для матрицы:\n{m1}")
```

```python
print(f"детерминант второй матрицы:\n{m2}\n равен {det2}")
```

```python
inv2 = np.linalg.inv(m2)
print(f"Обратная матрица для матрицы:\n{m2}\n\nравна:\n{inv2}")
```

##### 17. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/2023_new/ML0.1%20intro/data.csv")

df.Rooms = df.Rooms.astype("int8")


df.groupby("Rooms").count()
```

```python
rooms = df.groupby("Rooms")["Id"].count()

rooms
```

```python
rooms = df.groupby("Rooms")["Id"].count()


# Создать массив, в котором сектор с наибольшим числом квартир выдвинут
explode = [0] * len(rooms)
explode[rooms.argmax()] = 0.1


plt.figure(figsize=(30, 30))
rooms.plot(kind="pie", y="Price", autopct='%1.1f%%', explode=explode)

plt.ylabel("")
plt.title("Процент квартир по количеству комнат")


# # Все значения кроме 1, 2, 3, 4 имеют слишком маленькую долю и поэтому сливаются в одну точку
# # Как вариант можно объединить их и одну долю
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/2023_new/ML0.1%20intro/data.csv")
df.Rooms = df.Rooms.astype("int8")

room_counts = df.groupby(["Rooms"]).count()

explode = [0] * len(rooms)
explode[rooms.argmax()] = 0.1

room_counts.plot(kind="pie", y="Id", autopct='%1.1f%%', explode=explode)


plt.ylabel("")
plt.title("Количество квартир по количеству комнат")

# Отображение диаграммы
plt.show()
```

```python
df["Preprocessing Rooms"] = df["Rooms"].apply(lambda x: x if x in [1, 2, 3, 4] else "5 and more")

prooms = df.groupby("Preprocessing Rooms")["Id"].count()

explode = [0] * len(prooms)
explode[prooms.argmax()] = 0.1

plt.figure(figsize=(15, 15))
prooms.plot(kind="pie", y="Price", autopct='%1.1f%%', explode=explode)

plt.ylabel("")
plt.title("Процент квартир по количеству комнат")
```
