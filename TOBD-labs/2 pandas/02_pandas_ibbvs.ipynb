```python
def pretty_out(dframe):
    columns_name = dframe.isna().sum().index
    columns_value = dframe.isna().sum().values
    
    return {k:v for k, v in zip(columns_name, columns_value)}

dict_recipes = pretty_out(recipes)
length_recipes = recipes.shape[0]

dict_reviews = pretty_out(reviews)
length_reviews = reviews.shape[0]
print('recipes:')
[print(f"{key:>15}: {round(value/length_recipes * 100, 2) if value else value}%") for key, value in dict_recipes.items()]

print('\n\n\nreviews:')
[print(f"{key:>15}: {round(value/length_recipes * 100, 2) if value else value}%") for key, value in dict_reviews.items()]
```
